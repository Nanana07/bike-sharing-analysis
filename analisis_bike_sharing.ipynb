## Visualisasi Distribusi Suhu
Di bawah ini adalah histogram yang menggambarkan distribusi suhu pada dataset. Ini akan membantu kita memahami rentang suhu yang sering muncul dalam data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca dataset
import pandas as pd
df = pd.read_csv("C:/Users/hp/Documents/Nanda/day.csv")

# Membuat histogram untuk suhu (temperature)
plt.figure(figsize=(10, 6))
sns.histplot(df['temp'], kde=True, color='blue')  # 'temp' adalah kolom suhu
plt.title('Distribusi Suhu pada Dataset')
plt.xlabel('Suhu')
plt.ylabel('Frekuensi')
plt.show()
```

## Korelasi antara Suhu dan Jumlah Pengguna
Di bawah ini adalah scatter plot yang menunjukkan hubungan antara suhu (`temp`) dan jumlah pengguna sepeda (`cnt`). Plot ini akan membantu kita melihat apakah suhu memiliki pengaruh terhadap jumlah sepeda yang digunakan.

```python
# Membuat scatter plot antara suhu dan jumlah pengguna sepeda
sns.scatterplot(x=df['temp'], y=df['cnt'])
plt.title('Korelasi antara Suhu dan Jumlah Pengguna Sepeda')
plt.xlabel('Suhu')
plt.ylabel('Jumlah Pengguna')
plt.show()
```

## Penggunaan Sepeda Berdasarkan Jam
Di bawah ini adalah bar plot yang menggambarkan jumlah pengguna sepeda berdasarkan jam (`hr`). Ini akan memberi kita gambaran waktu puncak peminjaman sepeda.

```python
# Membuat bar plot untuk penggunaan sepeda berdasarkan jam
hourly_usage = df.groupby('hr')['cnt'].sum()
hourly_usage.plot(kind='bar')
plt.title('Penggunaan Sepeda Berdasarkan Jam')
plt.xlabel('Jam')
plt.ylabel('Jumlah Pengguna')
plt.show()
```

## Kesimpulan
1. **Suhu** memiliki pengaruh terhadap jumlah pengguna sepeda. Ketika suhu berada di kisaran 20-30 derajat Celsius, jumlah sepeda yang digunakan lebih banyak.
2. Waktu puncak penggunaan sepeda adalah **jam 17:00**, menunjukkan bahwa banyak orang menggunakan sepeda di sore hari.
